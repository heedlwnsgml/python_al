```python
# 코드 1.7: 반복 구조의 팩토리얼 함수

def fact_iter(n) :
    result = 1
    for k in range(2, n+1) :	# k: 2, ..., n
        result = result * k	# result에 반복적으로 곱함
    return result


# 코드 1.8: 순환 구조의 팩토리얼 함수

def fact_recur(n) :		# 순환적으로 구현한 factorial 함수
    if n == 1 : return 1		# 종료 조건: 순환을 멈추는 부분
    else :
        return n * fact_recur(n - 1)	# 자신을 순환적으로 호출

def main():
    number = int(input("정수를 입력하세요 : "))
    print(f"{number}! = {fact_iter(number)}")
    print(f"{number}! = {fact_recur(number)}")

if __name__ == "__main__":
    main()
```

```python
#fibonacci sequence에서 n번째 항의 값 계산하기
# 1. 반복 구조로 설계
def fibo_iter(n):
    if n<= 0:
        return 0
    elif n== 1:
        return 1

    prev = 0
    current = 1

    for _ in range(2, n+1):
        prev, current = current, prev + current

    return current

# 2. 재귀적 구조로 설계
def fibo_recur(n):
    if n<= 0:
        return 0
    elif n ==1:
        return 1
    else :
        return fibo_recur(n-1) + fibo_recur(n-2)

# test
def main():
    index = int(input("피보나치 수열에서 구할 항을 입력하세요 : "))
    print(f"{index}! = {fibo_iter(index)}")
    print(f"{index}! = {fibo_recur(index)}")

if __name__ == "__main__":
    main()
```

```python
# 원형 큐 구현: 배열 방식 (전역 변수 사용)

# 전역 변수 선언
array = []
front = 0
rear = 0
capacity = 0

def create_queue(c=10): # 큐의 생성 및 초기화
    global array, front, rear, capacity
    capacity = c #용량(크기) 설정
    array = [None] * capacity  # 순차구조 : 용량에 맞는 (None을 초기값으로 한)배열 생성
    front = 0  # 전단 인덱스 초기화
    rear = 0   # 후단 인덱스 초기화

def is_empty(): # 공백상태
    global front, rear
    return front == rear

def is_full(): # 포화상태
    global front, rear, capacity
    return front == (rear + 1) % capacity  # 원형 큐에서는 후단이 전단 바로 앞에 있으면 포화 상태

def enqueue(item):
    global array, rear, front, capacity
    if not is_full():  # 포화 상태가 아닌 경우
        rear = (rear + 1) % capacity  # 후단 인덱스를 순환하도록 처리
        array[rear] = item
    else:
        print("Queue is full. Cannot enqueue.")

def dequeue():
    global array, front, rear, capacity
    if not is_empty():  # 공백 상태가 아닌 경우
      front = (front + 1) % capacity  # 전단 인덱스를 순환하도록 처리
      item = array[front]
      array[front] = None # 삭제된 자리 초기화
      return item
    else:
      print("Queue is empty. Cannot dequeue.")
      return None

def peek():
    global array, front, capacity
    if not is_empty():  # 공백 상태가 아닌 경우
        return array[front+1]
    else:
        print("Queue is empty. Nothing to peek.")
        return None

def size():
    global front, rear, capacity # 처음, 끝, 용량(크기)
    return (rear - front + capacity) % capacity  # 원형 큐에서 요소의 개수를 계산

def display(msg='Queue:'):
    global array, front, rear, capacity
    print(msg, end=' = [') 
    i = front
    while i != rear + 1:
        print(array[i], end=' ')
        i = (i + 1) % capacity  # 인덱스를 순환하도록 처리
    print(']')

# main 함수를 통해 동작 테스트
def main():
    # 큐 생성
    create_queue(5)  # 용량 5인 원형 큐 생성
    print("큐가 생성되었습니다.")
    display()
    print(f"{rear}, {front}")

    # 요소 삽입 (enqueue)
    print("\n요소 삽입 (enqueue) 작업을 수행합니다.")
    enqueue(10)
    print(f"{rear}, {front}")
    enqueue(20)
    print(f"{rear}, {front}")
    enqueue(30)
    print(f"{rear}, {front}")
    display()

    # 큐의 크기 확인
    print("\n큐의 현재 크기:", size()) # 3

    # 요소 삭제 (dequeue) - front안 10 삭제
    print("\n요소 삭제 (dequeue) 작업을 수행합니다.")
    dequeue()
    display()

    # 큐의 상단 요소 확인 (peek)
    print("\n상단 요소 확인 (peek):", peek()) #20

    # 요소 추가 삽입 (enqueue)
    print("\n다시 요소 삽입 (enqueue) 작업을 수행합니다.")
    enqueue(40)
    enqueue(50)
    enqueue(60)  # 큐가 포화 상태일 경우 메시지 출력
    display()

    # 큐가 포화 상태인지 확인
    print("\n큐가 포화 상태인가?:", is_full())

    # 큐의 크기 확인
    print("\n큐의 현재 크기:", size())

    # 모든 요소 삭제
    print("\n모든 요소를 삭제합니다.")
    dequeue()
    dequeue()
    dequeue()
    dequeue()  # 큐가 공백 상태일 경우 메시지 출력
    display()

    # 큐가 공백 상태인지 확인
    print("\n큐가 공백 상태인가?:", is_empty())

# main 함수 호출
if __name__ == "__main__":
    main()
```

```python
# Class를 이용한 배열 기반 원형큐 설계

class ArrayQueue:
    def __init__(self, capacity = 10):
        self.capacity = capacity
        self.array = [None] * capacity
        self.front = 0
        self.rear = 0

    def is_empty(self): # 공백상태
        return self.front == self.rear

    def is_full(self): # 포화상태
        return self.front == (self.rear + 1) % self.capacity  # 원형 큐에서는 후단이 전단 바로 앞에 있으면 포화 상태

    def enqueue(self, item):
        if not self.is_full():  # 포화 상태가 아닌 경우
            self.rear = (self.rear + 1) % self.capacity  # 후단 인덱스를 순환하도록 처리
            self.array[self.rear] = item
        else:
            print("array is full. Cannot enqueue.")

    def dequeue(self):
        if not self.is_empty():  # 공백 상태가 아닌 경우
          self.front = (self.front + 1) % self.capacity  # 전단 인덱스를 순환하도록 처리
          item = self.array[self.front]
          self.array[self.front] = None # 삭제된 자리 초기화
          return item
        else:
          print("array is empty. Cannot dequeue.")
          return None
    
    def peek(self):
        if not self.is_empty():  # 공백 상태가 아닌 경우
            return self.array[(self.front+1) % self.capacity]
        else:
            print("array is empty. Nothing to peek.")
            return None
    
    def size(self):
        return (self.rear - self.front + self.capacity) % self.capacity  # 원형 큐에서 요소의 개수를 계산
    
    def display(self, msg='Queue:'):
        print(msg, end=' = [') 
        
        i = self.front
        while i != self.rear:
            print(self.array[i], end=' ')
            i = (i + 1) % self.capacity  # 인덱스를 순환하도록 처리
        print(']')

# 테스트 프로그램
import random
if __name__ == "__main__":
    #1. create Queue
    q = ArrayQueue(8)
    #2. print q
    q.display("초기상태 큐 출력")
    #3. 원형 큐에 랜덤 데이터를 생성(0 ~ 99 사이의 정수)해서 저장
    while not q.is_full():
        q.enqueue(random.randint(0,100))
        print(q.array)
        print(q.rear)
        print(q.front)
    #4. 큐의 포화 상태를 출력
    q.display("포화 상태")
    #5. 큐의 데이터 삭제해서 출력
    print("삭제 순서 : ", end ='')
    while not q.is_empty():
        print(q.dequeue())
        print(q.array)
        print(q.rear)
        print(q.front)
    
```

```python
class ArrayQueue:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.array = [None] * capacity
        self.front = 0
        self.rear = 0

    def is_empty(self):  # 공백상태
        return self.front == self.rear

    def is_full(self):  # 포화상태
        return self.front == (self.rear + 1) % self.capacity  # 원형 큐에서는 후단이 전단 바로 앞에 있으면 포화 상태

    def enqueue(self, item):
        if not self.is_full():  # 포화 상태가 아닌 경우
            self.rear = (self.rear + 1) % self.capacity  # 후단 인덱스를 순환하도록 처리
            self.array[self.rear] = item  # 현재 rear 위치에 아이템 추가
            #위의 두 코드의 위치만 바꾸면 코드 실행문의 첫 번째에 위치한 None을 없앨 수 있다.
        else:
            print("array is full. Cannot enqueue.")

    def dequeue(self):
        if not self.is_empty():  # 공백 상태가 아닌 경우
            self.front = (self.front + 1) % self.capacity  # 전단 인덱스를 순환하도록 처리
            item = self.array[self.front]
            self.array[self.front] = None  # 삭제된 자리 초기화
            return item
        else:
            print("array is empty. Cannot dequeue.")
            return None

    def peek(self):
        if not self.is_empty():  # 공백 상태가 아닌 경우
            return self.array[(self.front + 1) % self.capacity]  # 다음 삭제할 요소를 반환
        else:
            print("array is empty. Nothing to peek.")
            return None

    def size(self):
        return (self.rear - self.front + self.capacity) % self.capacity  # 원형 큐에서 요소의 개수를 계산

    def display(self, msg='Queue:'):
        print(msg, end=' = [')
        i = self.front
        while i != self.rear:
            print(self.array[i], end=' ')
            i = (i + 1) % self.capacity  # 인덱스를 순환하도록 처리
        print(']')
# 테스트 프로그램
import random
if __name__ == "__main__":
    #1. create Queue
    q = ArrayQueue(8)
    #2. print q
    q.display("초기상태 큐 출력")
    #3. 원형 큐에 랜덤 데이터를 생성(0 ~ 99 사이의 정수)해서 저장
    while not q.is_full():
        q.enqueue(random.randint(0,100))
        print(q.array)
        print(q.rear)
        print(q.front)
    #4. 큐의 포화 상태를 출력
    q.display("포화 상태")
    #5. 큐의 데이터 삭제해서 출력
    print("삭제 순서 : ", end ='')
    while not q.is_empty():
        print(q.dequeue())
        print(q.array)
        print(f"rear: {q.rear}, front: {q.front}")
    
```
